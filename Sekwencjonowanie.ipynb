```python
from random import randint
from math import ceil

from Bio import SeqIO
```

```python
# more than one records, so I need to use parse,
# and next to get the first one
Y = next(SeqIO.parse("hs_ref_GRCh38.p2_chrY.fa", "fasta"))
Y
```

```python
len(Y.seq)
```

```python
DNAlength = len(Y.seq)
readLength = 150
noOfReads = ceil(DNAlength/readLength*10)
noOfReads
```

```python
reads =[]
for i in range(noOfReads):
    randN = randint(0,DNAlength-readLength-1)
    reads.append(Y.seq[randN:randN+readLength])
```
