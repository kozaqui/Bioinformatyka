# BLAST w Biopythonie
Musimy podać jakiego typu BLASTA chcemy użyć, bazę, względem której będziemy przeszukiwać, oraz sekwencję, względnie numer identyfikacyjny GI:

```python
import os
import statistics
from Bio.Blast import NCBIWWW
from Bio.Blast import NCBIXML
from Bio import SeqIO
```

```python
# print current working directory
print(os.getcwd())
print()

help(NCBIWWW.qblast)
```

```python
result_handle = NCBIWWW.qblast("blastn", "nr", "23527284")
print(result_handle.read())
```

```python
# zmniejszanie liczby wyników za pomocą hitlist_size
s = SeqIO.read("lab5-seq1.fasta", format="fasta")
result_handle2 = NCBIWWW.qblast("blastn", "nt", s.seq, hitlist_size=1)
```

Podstawowe bazy:
- nr - sekwencje białkowe
- nt - nukleotydowe

inne pod linkiem:
[dokumentacja BLASTa](blast.ncbi.nlm.nih.gov/Blast.cgi?CMD=Web&PAGE_TYPE= BlastDocs&DOC_TYPE =ProgSelectionGuide#db$)$)

```python
blast_records = NCBIXML.parse(result_handle2)
print(blast_records)
```

## Python 2 / Python 3 alert
W Pythonie 3 nie ma metody **something.next()**, tylko funkcja **next(something)** !!

```python
blast_record = next(blast_records)
print(blast_record)
```

>Blast being a local alignment algorithm, it can create several hits for the same query and subject to avoid using a large number of consecutive gaps.  Each of those hits are called HSP, standing for High Scoring Pair.
[Source](http://www.genomequest.com/docs/blast-hsp/)

```python
for alignment in blast_record.alignments:
    print('Alignment------------------------------')
    print('title: ', alignment.title)
    print('length: ', alignment.length)
    for hsp in alignment.hsps:
        print('HSP : ')
        print('e value:', hsp.expect)
        print(hsp.query[0:75] + '...')
        print(hsp.match[0:75] + '...')
        print(hsp.sbjct[0:75] + '...')
```

# Zadanie na punkty
Dla białka o sekwencji z zadania 1.5 (lab5-seq2.fasta) wyszukaj za pomocą Bio.Blast 50 pierwszych dopasowań. Wypisz dla nich tytuł dopasowanej sekwencji oraz długość dopasowania. Oblicz i wypisz średnią długość dopasowania oraz średnią wartość E dla tych sekwencji.

```python
s = SeqIO.read("lab5-seq2.fasta", format="fasta")
result_handle3 = NCBIWWW.qblast("blastp", "nr", s.seq, hitlist_size=50)
```

```python
blast_records = NCBIXML.parse(result_handle3)
blast_record = next(blast_records)
```

```python
for alignment in blast_record.alignments:
    print('\n====================== Alignment ======================')
    print('title: ', alignment.title)
    print('length: ', alignment.length)
```

```python
alignment_lengths = []
alignment_e_values = []
for alignment in blast_record.alignments:
    alignment_lengths.append(alignment.length)
    for hsp in alignment.hsps:
        alignment_e_values.append(hsp.expect)
```

```python
print(alignment_lengths)
print(alignment_e_values)
```

```python
print('Średnia długość dopasowania: ',statistics.mean(alignment_lengths))
print('Średnia wartość E, dla sekwencji wynikowych: ', statistics.mean(alignment_e_values))
```
